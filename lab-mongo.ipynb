```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Para trabajar con datos de tipo fecha
# ------------------------------------------------------------------------------
from datetime import datetime

# Para trabajar con Mongo
# ------------------------------------------------------------------------------
import pymongo

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
# lo primero que tenemos que hacer es conectarnos con Mongo
cliente = pymongo.MongoClient()
print(cliente)

# como la conexión se ha realizado con éxito vamos a ver las bases de datos que tenemos
cliente.list_database_names()
```

```python
# nos conectamos con una BBDD 
companies = cliente['Companies']

#Veamos además las colecciones de la BBDD, usando el método 'list_collection_names()'
companies.list_collection_names()
```

```python
compa = companies['Company']
```

```python
df = pd.DataFrame(compa.find())
df.head(2)
```

### 1. Todas las empresas cuyo nombre sea 'Babelgum'. Recupera solo su campo `name`.


```python
name_df = pd.DataFrame(compa.find({'name':'Babelgum'},{'_id':0,'name':1}))
name_df
```

### 2. Todas las empresas que tienen más de 5000 empleados. Limita la búsqueda a 20 empresas y ordénalas por **número de empleados**.


```python
employee_df = pd.DataFrame(compa.find({'number_of_employees':{'$gt':5000}}).limit(20).sort('number_of_employees',-1))
employee_df.head(1)
```

```python
len(employee_df)
```

### 3. Todas las empresas fundadas entre 2000 y 2005, incluyendo ambos años. Recupera solo los campos `name` y `founded_year`.


```python
year_df = pd.DataFrame(compa.find({'$and':[{'founded_year':{'$gte':2000}},#Mayor igual 2000
                                           {'founded_year':{'$lte':2005}}]},#Menor igual 2005
                                           {'_id':0,'name':1,'founded_year':1}))#Mostrar name y founded solo
year_df
```

### 4. Todas las empresas que tuvieron una valoración de más de 100.000.000 y que fueron fundadas antes de 2010. Recupera solo los campos `name` e `ipo`.


```python
rating_df = pd.DataFrame(compa.find({'$and':[{'ipo.valuation_amount':{'$gt':100000000}},#Mayor igual 100000000
                                           {'founded_year':{'$lt':2010}}]},#Menor igual 2010
                                           {'_id':0,'name':1,'ipo':1}))#Mostrar name y founded solo
rating_df1 = rating_df["name"]
rating_df2 = rating_df["ipo"].apply(pd.Series)
rating_df = pd.concat([rating_df1, rating_df2], axis=1)
# Renombrar las columnas correctamente
rating_df.rename(columns={0: "name",1: "valuation_amount",2: "valuation_currency_code",
                          3: "pub_year",4: "pub_month",5: "pub_day",6: "stock_symbol"}, inplace=True)
rating_df.head()
```

### 5. Todas las empresas que tienen menos de 1000 empleados y que fueron fundadas antes de 2005. Ordénalas por el número de empleados y limita la búsqueda a 10 empresas.


```python
employee_df = pd.DataFrame(compa.find({'$and':[{'number_of_employees':{'$lt':1000}},#Menor que 1000
                                           {'founded_year':{'$lt':2005}}]}).limit(10).sort('number_of_employees',1))
employee_df.head(1)
```

### 6. Todas las empresas que no incluyen el campo `partners`.


```python
partners_df = pd.DataFrame(compa.find({"partners":{"$nin":["Company"]}}))
partners_df.head(1)W
```

### 7. Todas las empresas que tienen un valor nulo en el campo `category_code`.


```python
category_df = pd.DataFrame(compa.find({"category_code":None}))
category_df.head(1)
```

### 8. Todas las empresas que tienen al menos 100 empleados pero menos de 1000. Recupera solo los campos `name` y `number of employees`.


```python
employee2_df = pd.DataFrame(compa.find({'$and':[{'number_of_employees':{'$gte':100}},#Menor que 1000
                                           {'number_of_employees':{'$lt':1000}}]},{'_id':0,'name':1,'number_of_employees':1}))

employee2_df
```

### 9. Ordena todas las empresas por su precio de IPO en orden descendente.


```python
ipo_df = pd.DataFrame(compa.find().sort("ipo.valuation_amount",-1))
ipo_df.head(1)
```

### 10. Recupera las 10 empresas con más empleados, ordenadas por el `número de empleados`.


```python
employee3_df = pd.DataFrame(compa.find().sort("number_of_employees",-1).limit(10))
employee3_df.head(1)
```

### 11. Todas las empresas fundadas en el segundo semestre del año. Limita tu búsqueda a 1000 empresas.


```python
company_df = pd.DataFrame(compa.find({'$and':[{'founded_month':{'$gte':6}},#Menor que 1000
                                           {'founded_month':{'$lte':12}}]}).limit(1000))

company_df.head(1)
```

### 12. Todas las empresas fundadas antes del año 2000 que tienen un monto de adquisición superior a 10.000.000.


```python
company2_df = pd.DataFrame(compa.find({'$and':[{'founded_year':{'$lt':2000}},
                                           {'acquisition.price_amount':{'$gt':10000000}}]}))

company2_df.head(1)
```

### 13. Todas las empresas que han sido adquiridas después de 2010, ordénalas por el monto de adquisición, y recupera solo los campos `name` y `acquisition`.


```python
company2_df = pd.DataFrame(compa.find({'acquisition.acquired_year':{'$gt':2010}},{"_id":0,"name":1,"acquisition":1}))
company2_df.head(1)
```

### 14. Ordena las empresas por su `founded year`, recuperando solo los campos `name` y `founded year`.


```python
company3_df = pd.DataFrame(compa.find({'acquisition.acquired_year':{'$gt':2010}},{"_id":0,"name":1,"acquisition":1}))
company3_df.head(1)
```

### 15. Todas las empresas que han sido fundadas en los primeros siete días del mes, incluyendo el séptimo. Ordénalas por su `acquisition price` en orden descendente. Limita la búsqueda a 10 documentos.


```python
company4_df = pd.DataFrame(compa.find({'founded_day':{'$lte':7}}).sort("acquisition.price_amount",-1).limit(10))
company4_df.head(1)
```

### 16. Todas las empresas en la categoría 'web' que tienen más de 4000 empleados. Ordénalas por la cantidad de empleados en orden ascendente.


```python
company5_df = pd.DataFrame(compa.find({'$and':[{'category_code':{'$eq':"web"}},
                                           {'number_of_employees':{'$gt':4000}}]}).sort('number_of_employees',-1))
company5_df.head(2)
```

### 17. Todas las empresas cuyo monto de adquisición es superior a 10.000.000, y cuya moneda es 'EUR'.


```python
company6_df = pd.DataFrame(compa.find({'$and':[{'acquisition.price_amount':{'$gt':10000000}},
                                           {'acquisition.price_currency_code':{'$eq':"EUR"}}]}))
company6_df.head(2)
```
